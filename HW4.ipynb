18. Describe an algorithm that locates the last occurrence of the smallest element in a finite list of integers, where the integers in the list are not necessarily distinct.     

A simple algorithm could solve this problem. First I would initialize a variable named min_val and store the zeroth element of the list. I would initialize another variable named min_index and store $0$ there, i.e. the index of min_val. Then I would loop through the list from the first (second in non-cs terms) element to the $n_{th}$ element. Each time looping through I would check to see if the element at the index is smaller or equal to the min_val. If it is, then I will save the element at the current index to min_val and save the current index to min_index. At the end of the loop I will then return min_index. The time complexity of this algorithm is O(n). 

Here is a procedure doing what is described above:

``` 
procedure min_index(a_0, a_1, ..., a_n: finite list of integers):     
    min_val := a_0 {Store the zeroth value as the minimum value of the list}     
    min_index := 0 {Store the index of the zeroth value}      

    for i := 1 to n {n is the length of the list}   
        if element at i less than or equal to min_val:
            then min_val := element at i, min_index := i

    return min_index


```


20. Show that if there were a coin worth 12 cents, the cashierâ€™s algorithm using quarters, 12-cent coins, dimes, nickels, and pennies would not always produce change using the fewest coins possible.

Take the change needed to be given out to be 15 cents. If the cashiers algorithm was used it would choose 12-cents (the highest value possible that fits in) and then it would have to pick 3 pennies, for a total of 4 coins. The optimal choice to minimize the amount of coins used would have been a dime and a nickel. Since the cashiers algorithm cannot look ahead, it chooses an inoptimal path. Change given out with 12 cents = 4, while change given out without 12 cents is 2 in this case. Another example of a coin system with 12 cents solving the problem inoptimally is when change needed to be given out is 21 cents. 

### Would a standard coin system or one with a 12-cent coin minimize the amount of change needed to be given on average if both were to use the cashiers algorithm

````python
def cash_algo_w_12(change, i):
    x = [25, 12, 10, 5, 1]
    num_coins = 0
    if i % 10000 == 0:
        print(f'Change with 12')
    index = 0
    while change != 0:
        if i % 10000 == 0:
            print(f'  Iteration: {index}: {change}')
        if change - x[0] >= 0:
            change = change - x[0]
            
        elif change - x[1] >= 0:
            change = change - x[1]
            
        elif change - x[2] >= 0:
            change = change - x[2]
            
        elif change - x[3] >= 0:
            change = change - x[3]
            
        elif change - x[4] >= 0:
            change = change - x[4]
        
            
        num_coins = num_coins + 1
        index += 1
    return num_coins    
            
def cash_algo_wo_12(change, i):
    x = [25, 10, 5, 1]
    num_coins = 0
    if i % 10000 == 0:
        print(f'Change with 12')
    index = 0
    while change != 0:
        if i % 10000 == 0:
            print(f'  Iteration: {index}: {change}')
        if change - x[0] >= 0:
            change = change - x[0]
        elif change - x[1] >= 0:
            change = change - x[1]
        elif change - x[2] >= 0:
            change = change - x[2]
        elif change - x[3] >= 0:
            change = change - x[3]
        
        num_coins = num_coins + 1
        index += 1
    return num_coins
    
    
    
````

````python

import random
results_w_12 = []
results_wo_12 = []
for i in range(100000):
   
    x = random.randint(1, 100)
    if i % 10000 == 0:
        print(f'Starting change for both algorithms: {x}')
        
    _result_wo_12 = cash_algo_wo_12(x, i)
    _result_w_12 = cash_algo_w_12(x, i)
    if i % 10000 == 0:
        print("*******************************************")
    results_w_12.append(_result_w_12)
    results_wo_12.append(_result_wo_12)


    
    
````

````python
import numpy as np

w12 = np.array(results_w_12)
wo12 = np.array(results_wo_12)

print(f'Average number of coins needed to fulfill change between 1 and 100 with a 12 cent coin: {w12.mean()}')
print(f'Average number of coins needed to fulfill change between 1 and 100 with standard coins only: {wo12.mean()}')

````
